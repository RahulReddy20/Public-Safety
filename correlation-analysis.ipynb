```python
df = pd.read_csv('/kaggle/input/dallas-public-safety/Police_Incidents.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import json
```

```python
# Load the dictionary from the JSON file
with open('unique_incidents.json', 'r') as f:
    unique_incidents_dict = json.load(f)

# Map 'Type of Incident' to 'Incident_Score' using the dictionary
df['Incident_Score'] = df['Type of Incident'].map(unique_incidents_dict)

# Display the new DataFrame with 'Incident_Score' added
print(df)
```

```python
# Define time bins
time_bins = {
    'Night': ('18:00', '23:59'),
    'Early Morning': ('00:00', '05:59'),
    'Morning': ('06:00', '11:59'),
    'Afternoon': ('12:00', '17:59')
}

# Create a function to assign time bins
def assign_time_bin(time):
    for bin_label, (start_time, end_time) in time_bins.items():
        if start_time <= time <= end_time:
            return bin_label
    return 'Unknown'

# Assign time bins to each row in the DataFrame
df['Time Bin'] = df['Time1 of Occurrence'].apply(assign_time_bin)
```

```python
# Define the ranges and corresponding labels for 'Incident_Score'
score_ranges = [(0, 10), (10, 30), (30, 60), (60, 80), (80, 100)]  # Wider score ranges
score_labels = ['Very Safe', 'Safe', 'Moderately Safe', 'Moderately Unsafe', 'Very Unsafe']
# Create a function to assign the label based on the score
def assign_safety_score(score):
    for i, (start, end) in enumerate(score_ranges):
        if start <= score < end:
            return score_labels[i]
    return None  # Handle values outside the defined ranges
# Apply the function to create the 'Safety Score' column
df['Safety Score'] = df['Incident_Score'].apply(assign_safety_score)
```

```python
df = df.dropna(subset=['Zip Code', 'Safety Score', 'Incident_Score'])
```

```python
# Calculate the number of NaN values in each column
nan_counts = df.isnull().sum()
# Print the column names and their counts of NaN values
for column, count in nan_counts.items():
    if count == 0:
        print(f"Column: {column}, Count of NaN: {count}")
```

```python
# Get the column names with zero NaN values
columns_with_no_nan = nan_counts[nan_counts == 0].index

# Create a DataFrame containing only the columns with no NaN values
df_no_nan = df[columns_with_no_nan]

# Get the data types of columns with no NaN values
column_data_types = df_no_nan.dtypes

# Create a new DataFrame for displaying column names and data types
data_types_table = pd.DataFrame({'Column Name': column_data_types.index, 'Data Type': column_data_types.values})

```
